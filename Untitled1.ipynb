```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import numpy as np

# 1. 加载auto - mpg数据集
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
col_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration','model_year', 'origin', 'car_name']
# 使用sep='\s+'替代delim_whitespace=True
df = pd.read_csv(url, sep='\s+', names=col_names)

# 检查并处理非数值数据，这里简单将无法转换为数值的'?'替换为NaN
for col in ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 2. 选择连续字段作为特征
continuous_cols = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
X = df[continuous_cols]

# 3. 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=continuous_cols)

# 4. 处理缺失值，用均值插补
X_scaled = X_scaled.fillna(X_scaled.mean())

# 5. 执行层次聚类
agg_cluster = AgglomerativeClustering(n_clusters = 3, linkage='average')
X_scaled['cluster'] = agg_cluster.fit_predict(X_scaled)

# 6. 获取每个聚类的均值和方差
cluster_stats = X_scaled.groupby('cluster').agg(['mean', 'var'])
print("每个聚类的均值和方差:")
print(cluster_stats)

# 7. 以origin为类标签，获取每个类的均值和方差
X_scaled['origin'] = df['origin']
class_stats = X_scaled.groupby('origin').agg(['mean', 'var'])
print("\n以origin为类标签时每个类的均值和方差:")
print(class_stats)

# 8. 比较聚类分配和类标签之间的关系
# 这里简单通过观察数据判断，实际可能需要更复杂的统计检验
print("\n通过观察均值和方差，聚类分配和类标签之间没有非常明显的关系。聚类的均值和方差分布与以origin为类标签时的分布模式差异较大。")

# 9. 生成Cluster vs Origin的交叉表
cluster_vs_origin_crosstab = pd.crosstab(X_scaled['cluster'], X_scaled['origin'])
print("\nCluster vs Origin Crosstab:")
print(cluster_vs_origin_crosstab)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=X_scaled, x='mpg', y='weight', hue='cluster', palette='deep', s=100)
plt.title('Hierarchical Clustering Results (Auto-MPG Dataset)')
plt.xlabel('MPG (Standardized)')
plt.ylabel('Weight (Standardized)')
plt.legend(title='Cluster')
plt.savefig("mpg_scatter.png")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# Load Boston dataset
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
X = pd.DataFrame(data, columns=features)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features)

# Perform KMeans clustering for k from 2 to 6 and collect silhouette scores
silhouette_scores = []
k_range = range(2, 7)

print("Silhouette Scores for different k:")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42,n_init=10)
    kmeans.fit(X_scaled)
    score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"k = {k}: Silhouette Score = {score:.4f}")

# Find optimal k
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"\nOptimal k: {optimal_k}, Silhouette Score: {max(silhouette_scores):.4f}")

# Perform KMeans with optimal k
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_optimal.fit(X_scaled)
X['Cluster'] = kmeans_optimal.labels_

# Calculate mean of each feature per cluster
cluster_means = X.groupby('Cluster').mean()

# Get centroids
centroids = pd.DataFrame(scaler.inverse_transform(kmeans_optimal.cluster_centers_), 
                         columns=features)

# Print results
print("\nCluster Means:")
print(cluster_means)
print("\nCentroids:")
print(centroids)
```

```python

plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Scores for K-Means Clustering (Boston Dataset)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.savefig("silhouette_boston.png")
plt.show()
```

```python
# 进行 K-Means 聚类 (k=2)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
X['Cluster'] = kmeans.fit_predict(X_scaled)

# 获取质心
centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=features)

# 散点图
plt.figure(figsize=(8, 6))
sns.scatterplot(data=X, x='CRIM', y='LSTAT', hue='Cluster', palette='deep', s=100)
plt.scatter(centroids['CRIM'], centroids['LSTAT'], c='red', marker='x', s=200, label='Centroids')
plt.title('K-Means Clustering Results (Boston Dataset)')
plt.xlabel('Crime Rate')
plt.ylabel('Lower Status Population')
plt.legend()
plt.savefig("boston_scatter.png")
plt.show()
```
